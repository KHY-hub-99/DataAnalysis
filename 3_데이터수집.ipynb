# 1. 크롤링

크롤링(Crawling)이란 웹 크롤러(Web Crawler) 또는 스크레이퍼(Scraper)라고 불리는 프로그램이나 스크립트를 사용하여 인터넷상의 웹 페이지에서 데이터를 자동으로 수집하는 과정을 말합니다. 주로 검색 엔진이 웹사이트를 탐색하고 색인(Index)에 추가하기 위해 사용하는 기술이며, 특정 주제나 데이터를 수집해 분석하는 데에도 활용됩니다. 크롤링은 HTML, CSS, JavaScript로 구성된 웹페이지 구조를 파싱(Parsing)하여 원하는 정보를 추출하고, 이를 활용 가능한 데이터 형식으로 저장합니다. 크롤링 시에는 웹사이트의 이용 약관과 로봇 배제 표준(Robots.txt)을 준수하여 법적, 윤리적 문제를 방지해야 합니다.

# 2.basicenglishspeaking

https://basicenglishspeaking.com/daily-english-conversation-topics/

```python
import requests
```

```python
site = "https://basicenglishspeaking.com/daily-english-conversation-topics/"
req = requests.get(site)
print(req)
```

```python
from bs4 import BeautifulSoup as bs

soup = bs(req.text, "html.parser")
div = soup.find("div", {"class":"thrv-columns"})
a = div.find_all("a")

subject = []
for a in a:
  subject.append(a.text)

print(subject)
```

# 3. 다음 뉴스기사 제목 크롤링

https://v.daum.net/v/20251204120206933

```python
def daum_news_title(news_id):
  url = f"https://v.daum.net/v/{news_id}"
  req = requests.get(url)
  soup = bs(req.text, "html.parser")
  title = soup.find("h3", {"class":"tit_view"})
  if title:
    return title.text.strip()
  return "제목없음"
```

```python
daum_news_title(20251204120206933)
```

# 4. 벅스 차트 노래 제목 아티스트 가져오기

https://music.bugs.co.kr/chart

```python
url = "https://music.bugs.co.kr/chart"
req = requests.get(url)
soup = bs(req.text, "html.parser")
```

```python
titles = soup.find_all("p", {"class":"title"})
artists = soup.find_all("p", {"class":"artist"})
print(zip(titles, artists))
```

```python
for i, (t, a) in enumerate(zip(titles, artists)):
  title = t.text.strip()
  artist = a.text.strip().split("\n")[0]
  print(f"{i+1}. {title} - {artist}")
```

# 5. 멜론차트

https://www.melon.com/chart/index.htm

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36"}
req = requests.get("https://www.melon.com/chart/index.htm", headers=headers)
req
```

# 6. 네이버 증권

https://finance.naver.com/item/main.naver?code={}

```python
codes = [
    "005930",  # 삼성전자
    "000660",  # SK하이닉스
    "373220",  # LG에너지솔루션
    "207940",  # 삼성바이오로직스
    "005380",  # 현대자동차
    "000270",  # 기아
    "329180",  # 현대중공업 (HD현대중공업)
    "105560",  # KB금융
    "034020",  # 두산에너빌리티
    "068270",  # 셀트리온
]
```

```python
def finance(code):
  url = f"https://finance.naver.com/item/main.naver?code={code}"
  req = requests.get(url)
  soup = bs(req.text, "html.parser")
  name = soup.select_one("#middle > div.h_company > div.wrap_company > h2 > a").text
  code = soup.select_one(".code").text
  price = soup.select_one(".no_today span.blind").text
  price_int = int(price.replace(",", ""))
  volume = soup.select_one(".no_info tr:nth-of-type(1) td:nth-of-type(3) em > span.blind").text
  volume_int = int(volume.replace(",", ""))
  dic = {"name":name, "code":code, "price":price_int, "volume":volume_int}
  return dic
```

```python
data = []
for code in codes:
  data.append(finance(code))
```

```python
data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
df.sort_values(by="price", ascending=False)
```

```python
df.to_excel("naver_finance.xlsx")
```
