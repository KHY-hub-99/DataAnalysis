### ※ xpath
XPath는 XML 또는 HTML 문서 내에서 특정 요소나 속성을 선택하기 위해 사용되는 경로 표현 언어입니다. 웹 크롤링이나 자동화 도구에서 주로 사용되며, 요소를 효율적으로 찾을 수 있도록 도와줍니다. 일반적인 XPath는 특정 위치나 속성을 기준으로 요소를 선택하는 상대적인 경로를 사용합니다. 또한 full xpath는 루트 요소에서 시작하여 대상 요소까지의 절대적인 경로를 나타냅니다. 따라서 문서 구조가 변경되면 경로가 깨질 가능성이 높습니다.

/html/body/div[1]/div[5]/div/div/div[5]/div[1]/div[3]/div/section/ul/li[1]/div[3]/ul/li[1]


```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
driver.get("https://comic.naver.com/webtoon/detail?titleId=776601&no=178&week=fri")
xpath = "/html/body/div[1]/div[5]/div/div/div[5]/div[1]/div[3]/div/section/ul/li"
best_comment_elements = driver.find_elements(By.XPATH, xpath)
print(best_comment_elements)
```

```python
for li in best_comment_elements:
    try:
        comment_p = li.find_element(By.XPATH, "./div[1]/div[2]/div/p")
        comment_text = comment_p.text.strip()
        print("=================================")
        print(comment_text)
    except Exception as e:
        print(e)
```

```python
d = webdriver.Chrome()
d.get("https://comic.naver.com/webtoon/detail?titleId=776601&no=178&week=fri")
d.find_element(By.XPATH, "/html/body/div[1]/div[5]/div/div/div[5]/div[1]/div[3]/div/section/div[4]/button[2]").click()
time.sleep(2)
x = "/html/body/div[1]/div[5]/div/div/div[5]/div[1]/div[3]/div/section/ul/li"
cm = d.find_elements(By.XPATH, x)

print("******* 전체 댓글 *******")
for li in cm:
    try:
        comment_p = li.find_element(By.XPATH, "./div[1]/div[2]/div/p")
        comment_text = comment_p.text.strip()
        print("=================================")
        print(comment_text)
    except Exception as e:
        continue
```

# 3. 픽사베이 이미지 다운로드

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from urllib.request import Request, urlopen
```

```python
d = webdriver.Chrome()
url = "https://pixabay.com/ko/images/search/cat/"
d.maximize_window()
d.get(url)
```

```python
img_x = "/html/body/div[1]/div[1]/div/div[2]/div[1]/div[2]/div[2]/a/img"
wait = WebDriverWait(d, 10) # 최대 10초까지 기다림
img_e = wait.until(EC.presence_of_element_located((By.XPATH, img_x)))
```

```python
img_url = img_e.get_attribute("src")
img_byte = Request(img_url, headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36"})
img_byte
```

```python
with open("cat.jpg", "wb") as f:
    f.write(urlopen(img_byte).read())
print("File Save!")
d.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from urllib.request import Request, urlopen
import pandas as pd
import openpyxl
```

```python
def yanolza(name, url, max_scroll=3):
    review_list = []
    d = webdriver.Chrome()
    d.get(url)
    time.sleep(3)

    last_height = d.execute_script("return document.body.scrollHeight")
    # print(last_height)
    same_count = 0
    scroll_times = 0

    while True:
        scroll_times += 1
        d.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        time.sleep(2)
        new_height = d.execute_script("return document.body.scrollHeight")

        if new_height == last_height:
            same_count += 1
        else:
            same_count = 0
            last_height = new_height
        
        if same_count >= 3:
            print("더 이상 새로운 리뷰가 없습니다.")
            break
        
        if scroll_times >= max_scroll:
            print("최대 스크롤 횟수를 넘었습니다.")
            break

    containers = d.find_elements(By.XPATH, "/html/body/div[1]/section/div/div[5]/div")
    # print("총 리뷰 갯수 :", len(containers))
    for c in containers:
        try:
            # raw_text = c.text.strip()
            date_text = c.find_element(By.XPATH, "./div[2]/div/div[1]/div[1]/div[2]/p").text
            review_text = c.find_element(By.XPATH, "./div[2]/div/div[3]/div[1]/p").text
        except:
            raw_text = ""
        
        review_list.append({"review": review_text, "date":date_text})
    
    d.quit()
    return review_list
```
